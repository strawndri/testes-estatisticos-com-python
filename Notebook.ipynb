# 1 Importações

```python
import pandas as pd
import seaborn as sns
```

# 2 Conhecendo os dados

```python
tmdb = pd.read_csv('tmdb_5000_movies.csv')
tmdb.head()
```

```python
notas = pd.read_csv("ratings.csv")
notas.head()
```

# 3 Análise visual dos dados

## 3.1 TMDB

```python
ax = sns.displot(tmdb.vote_average, kde=False, bins=50)
ax.set(xlabel='Nota Média', ylabel='Frequência', title='Média de votos em filmes no TMDB 5000')
```

```python
ax = sns.boxplot(x=tmdb.vote_average)
ax.set(xlabel='Nota média do filme')
ax.set_title('Distribuição de nota média dos filmes do TMDB 5000')
```

```python
tmdb.query('vote_average == 0')
```

```python
tmdb_mais_10_votos = tmdb.query('vote_count >= 10')
tmdb_mais_10_votos.describe()
```

```python
ax = sns.displot(tmdb_mais_10_votos.vote_average, kde=False, bins=50)
ax.set(xlabel='Nota Média', ylabel='Frequência', title='Média de votos em filmes (com 10 ou mais votos) no TMDB 5000')
```

## 3.2 MovieLens

```python
nota_media_por_filme = notas.groupby("movieId").mean()["rating"]
nota_media_por_filme.head()
```

```python
ax = sns.displot(nota_media_por_filme.values)
ax.set(xlabel='Nota média', ylabel='Frequência', title='Média de votos em filmes no MovieLens')
```

```python
quantidade_de_votos_por_filme = notas.groupby("movieId").count()
quantidade_de_votos_por_filme.query("rating >= 10")
```

```python
quantidade_de_votos_por_filme = notas.groupby("movieId").count()
filmes_com_pelo_menos_10_votos = quantidade_de_votos_por_filme.query("rating >= 10").index
```

```python
nota_media_dos_filmes_com_pelo_menos_10_votos = nota_media_por_filme.loc[filmes_com_pelo_menos_10_votos.values]
nota_media_dos_filmes_com_pelo_menos_10_votos.head()
```

```python
ax = sns.displot(nota_media_dos_filmes_com_pelo_menos_10_votos)
ax.set(xlabel='Nota média', ylabel='Frequência', title='Média de votos em filmes no MovieLens')
```

```python
ax = sns.boxplot(x=nota_media_dos_filmes_com_pelo_menos_10_votos.values)
ax.set(xlabel='Nota  média do filme')
ax.set_title('Distribuição de nota média dos filmes do MovieLens')
```
